```python
x = np.array([[6,9,3,5,8],[6,7,9,8,8],[1,8,2,1,5],[1,3,6,2,9]])

print(x,"\n========")

print(f"Sum of rows        => {np.sum(x,axis=1)}")
print(f"Avg of cols        => {np.mean(x,axis=0)}")
print(f"Min vals in rows   => {np.min(x,axis=1)}")
print(f"Median of each col => {np.median(x,axis=0)}\n")
print(f"covariance of each col:\n{np.cov(x)}\n")
print(f"Correlation coefficient :\n{np.corrcoef(x)}")
print("---------")
print(f"Pos of largest in each row    => {np.argmax(x,axis=1)}")

print(f"All vals in cols are non_zeros => {np.all(x,axis=0)}")
x[0,1] = 0
print(f"All vals in cols are non_zeros => {np.all(x,axis=0)}")
```

```python
one_d_array = np.array([1,2,3,4,5,6])
two_d_array = np.array([one_d_array, one_d_array + 6, one_d_array + 12])

print(two_d_array,"\n")
```

```python
print(np.sum(two_d_array, axis=0))      # Get the col sums  
print(np.sum(two_d_array, axis=1),"\n") # Get the row sums 

print(np.mean(two_d_array))
print(np.mean(two_d_array,axis=0)) 
print(np.mean(two_d_array,axis=1),"\n") 

print(np.std(two_d_array))
print(np.std(two_d_array,axis=0)) 
print(np.std(two_d_array,axis=1),"\n") 

print(np.log(two_d_array),"\n")
print(np.sqrt(two_d_array))
```

```python
a = np.array([1, 2, 3])
b = np.array([5, 6, 7])
```

####  **Bitwise AND (& or np.bitwise_and)**

```python
result = np.bitwise_and(a, b)
print(result)  # Output: [1 2 3]
```

```python
## Explanation:
# Binary for 1: 0001
# Binary for 5: 0101
# 0001 & 0101 = 0001  1
```

####  **Bitwise OR (| or np.bitwise_or)**

```python
result = np.bitwise_or(a, b)
print(result)  # Output: [5 6 7]
```

```python
## Explanation:
# Binary for 1: 0001
# Binary for 5: 0101
# 0001 | 0101 = 0101  5
```

####  **Bitwise XOR (^ or np.bitwise_xor)**

```python
result = np.bitwise_xor(a, b)
print(result)  # Output: [4 4 4]
```

```python
## Explanation:
# Binary for 1: 0001
# Binary for 5: 0101
# 0001 ^ 0101 = 0100  4
```

####  **Bitwise NOT (~ or np.bitwise_not)**

```python
result = np.bitwise_not(a)
print(a)
print(result)  # Output: [-2 -3 -4]
```

```python
## Explanation:
# Binary for 1: 0001
# Negation of 0001 is 1110 which is -2 in two's complement representation.
```
