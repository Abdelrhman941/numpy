Arithmetic Operations

```python
x=np.array([1,3,4,2])
y=np.array([2,5,6,7])

print(np.mod(y,x))
print("----")
print(np.add(x,y))  
print("----")
print(np.subtract(x,y))  
print("----")
print(np.negative(x,y))  
print("----")
print(np.multiply(x,y))  
print("----")
print(np.divide(x,y))  
print("----")
print(np.floor_divide(x,y))  
```

```python
small_array1 = np.array([[1,2],[3,4]])
small_array2 = np.array([[2,3]])
small_array3 = np.array([[2,3],[1,2]])

print(small_array1)
print("---------")
print(small_array2)
print("---------")
print(small_array3)
print("---------")
print(small_array1 + small_array2,"\n")
print(small_array1 - small_array2,"\n")
print(small_array1** small_array2,"\n") # 1**2  2**3   3**2  4**2
print(small_array1** small_array3,"\n") # 1**2  2**3   3**1  4**2
```

> # **product** 

#### ðŸ‘‰ 1) `usual` :

```python
my_array3 = np.array([[1, 4], [5, 9]]) 
my_array4 = np.array([[2, 7], [10, 5]])
print(my_array3 * my_array4)
```

#### ðŸ‘‰ 2) `dot` :

```python
print(np.dot(small_array1, small_array1))
```

```python
one_d_array = np.array([1,2,3,4,5,6])
two_d_array = np.array([one_d_array, one_d_array + 6, one_d_array + 12])

print(two_d_array)
print("---------")
print(np.dot(two_d_array[0,:],two_d_array[1,:]),"\n") # Take the vector dot product of row 0 and row 1
```

<div style="background-color:rgb(85, 78, 79) ; color: #ffffff; width: 100%; height: 50px; text-align: center; font-weight: bold; line-height: 50px; margin: 10px 0; font-size: 24px;">
cum(sum-product)
</div>

```python
print(np.cumsum(two_d_array,axis=0))
print("------------------")
print(np.cumsum(two_d_array,axis=1))
```

```python
print(np.cumproduct(two_d_array,axis=0))
# [[row1          ]
#  [row1*row2     ]
#  [row1*row2*row3]]

print("---------")
print(np.cumproduct(two_d_array,axis=1))
# [[col1    col1*col2   col1*col2*col3  .......]
```

#### ðŸ‘‰ **`trace`** : add for the main diagonal

```python
a = np.arange(9) 
print(a)
print('-------')

b = a.reshape(3,3) 
print(b) 
print('-------')

c = np.trace(b) 
print(c)
```

#### ðŸ‘‰ **`linalg.det`** : find matrix value
#### ðŸ‘‰ **`linalg.eig`** : find eigenvalues & eigenvectors

```python
c = np.linalg.det(b)  
print(c)
print('-------')

d = np.linalg.eig(b) 
print(d) 
```

#### ðŸ‘‰ **`linalg.inv`** : find matrix inverse

```python
a = np.random.randint(1,4, size=9).reshape(3,3) 
print(a)
print('-------')
b = np.linalg.inv(a)        # sometimes it may not be possible to find the inverse of a matrix
print(b)
```
